I.E :- <, <=, >, >=, ==, !=,
PYTHON INTERNALLY COMPARES INDIVIDUAL ELEMENTS OF LIST IN LEXICOGRAPHICAL ORDER

```python
#PROGRAM TO COMPARE 2 LISTS
def compare(l1,l2):
    
    list1.sort()
    list2.sort()
    
    list1=[2,3,4,5]
    list2=[5,4,2,3]
    
    print("COMPARSION",compare(list1,list2))
    
    if(list1==list2):
        print("EQUAL LIST")
    else:
        print("NOT EQUAL LIST")
```

#07:- Write a Python program to count the number of strings where the string
length is 2 or more and the first and last character are same from a given
list of strings.

```python
def match_words(words):
  ctr = 0

  for word in words:
    if len(word) > 1 and word[0] == word[-1]:
      ctr += 1
  return ctr

print(match_words(['abc', 'xyz', 'aba', '1221']))
```

08:-Write a Python program to remove duplicates from a list.


```python
# Python to remove duplicate elements
def Remove(duplicate):
    final_list = []
    for num in duplicate:
        if num not in final_list:
            final_list.append(num)
    return final_list
     
# Driver Code
duplicate = [2, 4, 10, 20, 5, 2, 20, 4]
print(Remove(duplicate))
```

10:-Write a Python program to convert a list of characters into a string. 

```python
# Python program to convert a list
# of character

def convert(s):

	# initialization of string to ""
	new = ""

	# traverse in the string
	for x in s:
		new += x

	# return string
	return new
	
	
# driver code
s = ['a', 'b', 'd', 'u', 'l', 'l', 'a', 'h']
print(convert(s))
```
